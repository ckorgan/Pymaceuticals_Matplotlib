```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
# mouse_drug_data_to_load = "data/mouse_drug_data.csv"
# clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data

data1 = pd.read_csv('clinicaltrial_data.csv')
data2 = pd.read_csv('mouse_drug_data.csv')
# Combine the data into a single dataset

data = data2.merge(data1,how='inner')

# Display the data table for preview

data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_tumor_volume_data = data.groupby(by=['Drug','Timepoint'])['Tumor Volume (mm3)'].mean().reset_index()
# Convert to DataFrame
mean_tumor_volume_data = pd.DataFrame(mean_tumor_volume_data)
# Preview DataFrame
mean_tumor_volume_data
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
error_tumor_volume_data = data.groupby(['Drug','Timepoint'])['Tumor Volume (mm3)'].sem().reset_index()

error_tumor_volume_data = pd.DataFrame(error_tumor_volume_data)
error_tumor_volume_data.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
reshaped_data = mean_tumor_volume_data.pivot_table(index='Timepoint',columns=['Drug'],values='Tumor Volume (mm3)')

# Preview that Reformatting worked
reshaped_data.head()
```

```python
# Generate the Plot (with Error Bars)
plt.plot(reshaped_data.index,reshaped_data['Capomulin'],marker='o',color='red')
plt.plot(reshaped_data.index,reshaped_data['Infubinol'],marker='^',color='blue')
plt.plot(reshaped_data.index,reshaped_data['Ketapril'],marker='s',color='green')
plt.plot(reshaped_data.index,reshaped_data['Placebo'],marker='d',color='black')
plt.grid()
plt.xlabel('Time(Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.legend()
plt.title('Tumor Spread during Treatment')
# Save the Figure

plt.savefig('Time vs Tumor volume.png')
plt.show()
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_metastatic_sites = data.groupby(by=['Drug','Timepoint'])['Metastatic Sites'].mean()

# Convert to DataFrame
mean_metastatic_sites = pd.DataFrame(mean_metastatic_sites)
# Preview DataFrame
mean_metastatic_sites.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
error_metastatic_site = data.groupby(['Drug','Timepoint'])['Metastatic Sites'].sem()

# Convert to DataFrame
error_metastatic_site = pd.DataFrame(error_metastatic_site)

# Preview DataFrame
error_metastatic_site.head()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
survival = pd.DataFrame(data.groupby(by=['Drug','Timepoint'])['Metastatic Sites'].count().reset_index())
# Convert to DataFrame
survival['Mouse Count'] = survival['Metastatic Sites']
survival.drop(axis=1,labels=['Metastatic Sites'],inplace=True)
# Preview DataFrame
survival.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
reshaped_survival = survival.pivot_table(index='Timepoint',columns=['Drug'],values='Mouse Count')

# Preview the Data Frame
reshaped_survival.head()
```

```python
# Generate the Plot (Accounting for percentages)
percent_survival = reshaped_survival/25*100

# Generate the Plot (with Error Bars)
plt.plot(percent_survival.index,percent_survival['Capomulin'],marker='o',color='red')
plt.plot(percent_survival.index,percent_survival['Infubinol'],marker='^',color='blue')
plt.plot(percent_survival.index,percent_survival['Ketapril'],marker='s',color='green')
plt.plot(percent_survival.index,percent_survival['Placebo'],marker='d',color='black')
plt.grid()
plt.xlabel('Time(Days)')
plt.ylabel('Survival Rate(%)')
plt.legend()
plt.title('Survival During Treatment')
# Save the Figure

plt.savefig('Survival During Treatment.png')
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
# Display the data to confirm
percent_drug_success = reshaped_data/45*100
percent_change_drug = percent_drug_success.iloc[-1]-percent_drug_success.iloc[0]
percent_change_drug
```

```python
# Store all Relevant Percent Changes into a Tuple

percent_change_drug_tuple = list(zip((percent_change_drug).index, pd.DataFrame(percent_change_drug)[0]))
# Splice the data between passing and failing drugs

passing_drug=[]
failing_drug=[]
for i in percent_change_drug_tuple:
    if i[1]<0:
        failing_drug.append(i)
    else:
        passing_drug.append(i)

fig, ax = plt.subplots()    
rects=ax.bar(width=1,color=['green']+['red']*3,x=[failing_drug[0][0],passing_drug[1][0],passing_drug[2][0], passing_drug[4][0]], height=[failing_drug[0][1],passing_drug[1][1],passing_drug[2][1], passing_drug[4][1]],)
plt.ylabel('% Tumor Volume Change')
plt.title('Tumor Change Over 45 days Treatment');
# plt.bar()
plt.xticks([.5,1.5,2.5,3.5]);
plt.yticks(range(-20,60,20));
plt.grid()
plt.text(x=-0.2, y=-10, s='--19%',color='white')
plt.text(x=0.8, y=5, s='46%',color='white')
plt.text(x=1.8, y=5, s='57%',color='white')
plt.text(x=2.8, y=5, s='51%',color='white')

# Use functions to label the percentages of changes

# plt.text(x=10,va='center', text='12',color='white',s=10)
plt.savefig('Tumor Change Over 45 days Treatment.png')

plt.show()
```
